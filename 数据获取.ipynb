```python
import requests as rq
from bs4 import BeautifulSoup as bs
import re
```

```python
def gethtml(rooturl, encoding="utf-8"):
    # 默认解码方式utf-8
    response = rq.get(rooturl)
    response.encoding = encoding
    html = response.text
    return html  # 返回链接的html内容
```

```python
def getherf(html):
    # 使用BeautifulSoup函数解析传入的html
    soup = bs(html, features="lxml")
    allnode_of_a = soup.find_all("a")
    result = [_.get("href") for _ in allnode_of_a]
    # print(result)
    return result
```

```python
def filterurl(result):
    # result参数：get到的所有a标签内herf的内容
    # 对列表中每个元素进行筛选
    urlptn = r"http://www.pzhnews.org/qwfbhj/qxxw/(.+)"  # 匹配模式: 所有http://开头的链接
    urls = [re.match(urlptn, str(_)) for _ in result]  # 正则筛选
    # print(urls)
    while None in urls:
        urls.remove(None)  # 移除表中空元素
    urls = [_.group() for _ in urls]  # group方法获得re.match()返回值中的字符
    # print(urls)
    return urls
```

```python
link=[]
location=['dq','xq','rhq','my']
for lo in range(len(location)):
    for i in range(1,20):
        url="http://www.pzhnews.org/qwfbhj/qxxw/{0}/index_{1}.html".format(location[lo],i)
        html = gethtml(url)
        result = getherf(html)
        urls = filterurl(result)
        link+=urls
print(link)
```

```python
len(link)
```

```python
#超链接去重
news_link = []
for i in link:
    if i not in news_link:
        news_link.append(i)
print(news_link)
```

```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime
import json
import xlwt
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36"
}
title=[]
articleText=[]
article_all=[]
for i in news_link:
    url=i
    rep = requests.get(url, headers=headers)
    rep.encoding = 'utf-8'
    # print(rep.text) 输出的是网页的全部源代码
    html=BeautifulSoup(rep.text,'html.parser')
    #标题
    title.append(html.select('div.title')[0].text)
    #文本
    article=[]
    for p in html.select('div.xl-content p')[:-1]:
        article.append(p.text.strip())
        articleText=' '.join(article)
    article_all.append(articleText)
        
```

```python
book = xlwt.Workbook(encoding='utf-8')
sheet = book.add_sheet('攀枝花')
head = ['标题','内容','链接'] #表头
for h in range(len(head)):
    sheet.write(0,h,head[h]) #写入表头
for i in range(1,len(news_link)):
    try:
        sheet.write(i,0,title[i])
        sheet.write(i,1,article_all[i])
        sheet.write(i,2,news_link[i])
    except:
        print (str(i),'出现异常')
book.save('新闻.xls')            
```
